```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Análise de Dados PrograMaria/analise_dados - analise_dados.csv')
```

```python
dados.head()
```

```python
dados['IDADE'] = pd.to_numeric(dados['IDADE'], errors='coerce')
dados['SALARIO'] = pd.to_numeric(dados['SALARIO'], errors='coerce')
```

```python
genero_counts = dados['GENERO'].value_counts()
```

```python
plt.figure()
plt.bar(height = genero_counts.values, x = genero_counts.index)
plt.title('Quantidade de pessoas por gêneros na área de Dados')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.show()
```

```python
plt.figure()
sns.countplot(data=dados, x='GENERO', palette = 'pastel')
plt.title('Quantidade de pessoas por gêneros na área de Dados')
plt.show()
```

```python
salario_por_idade = dados.groupby('IDADE')['SALARIO'].mean()
salario_por_idade.head()
```

```python
plt.figure(figsize=(15, 4))
plt.plot(salario_por_idade.index, salario_por_idade.values, marker='o', linestyle='--')
plt.grid(True)
plt.title('Salário médio por idade')
plt.xlabel('Idade')
plt.ylabel('Salário médio')
plt.show()
```

```python
fig = px.line(salario_por_idade.reset_index(), x='IDADE', y='SALARIO', title='Média de salário por idade', markers=True)
fig.show()
```

```python
plt.figure(figsize=(15, 4))
plt.scatter(dados['IDADE'], dados['SALARIO'], alpha=0.5)
plt.title('Relação entre idade e salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.grid(True)
plt.show()
```

```python
fig = px.scatter(dados, x='IDADE', y='SALARIO', title='Relação entre idade e salário', trendline='ols')
fig.show()
```
